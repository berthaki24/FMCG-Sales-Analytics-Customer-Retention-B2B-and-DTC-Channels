```python
import pandas as pd

df = pd.read_csv('../data/sales_cleaned.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
# TẠO DIMENSION TABLES (Bảng Chiều)

# 1. Bảng Khách hàng (Dim_Customer)
# Drop duplicates để mỗi khách hàng chỉ xuất hiện 1 lần
dim_customer = df[['Customer ID', 'Customer Name', 'Sales_Channel', 'Country', 
                   'City', 'State']].drop_duplicates(subset=['Customer ID']).reset_index(drop=True)

# 2. Bảng Sản phẩm (Dim_Product)
dim_product = df[['Product ID', 'Product Name', 
                  'FMCG_Category']].drop_duplicates(subset=['Product ID']).reset_index(drop=True)

# 3. Bảng Thời gian (Dim_Date)
min_date = df['Order Date'].min()
max_date = df['Order Date'].max()
dim_date = pd.DataFrame({'Date': pd.date_range(min_date, max_date)})
dim_date['Year'] = dim_date['Date'].dt.year
dim_date['Month'] = dim_date['Date'].dt.month
dim_date['Quarter'] = dim_date['Date'].dt.quarter
dim_date['DayOfWeek'] = dim_date['Date'].dt.day_name()

# TẠO FACT TABLE (Bảng Sự kiện)

# Fact Table chỉ chứa khóa ngoại (Foreign Keys) và các biến số đo lường (Metrics)
fact_sales = df[['Order ID', 'Order Date', 'Customer ID', 'Product ID', 
                 'Sales', 'Quantity', 'Discount', 'Profit', 'Delivery_Time_Days', 
                 'Profit_Margin', 'High_Discount_Flag']]

# XUẤT DỮ LIỆU
dim_customer.to_csv('../data/Dim_Customer.csv', index=False)
dim_product.to_csv('../data/Dim_Product.csv', index=False)
dim_date.to_csv('../data/Dim_Date.csv', index=False)
fact_sales.to_csv('../data/Fact_Sales.csv', index=False)
```
